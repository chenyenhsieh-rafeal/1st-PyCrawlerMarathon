### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.ettoday.net/news/news-list.htm'

r=requests.get(url)

soup=BeautifulSoup(r.text,'html5lib')

title_list=[]

for title in soup.find(class_="part_list_2").find_all('h3'):
    title_list.append([title.find('span').text,title.find('em').text,title.find('a').text])
    print(title.find('span').text,title.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
columns=['時間','類別','標題']
category = pd.DataFrame(title_list, columns=columns)
category=category.groupby('類別').apply(lambda x: x.sort_values('類別', ascending=False))
category
```

### ③ 哪一個類別的文章最多

```python
category['類別'].value_counts()
```
