### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
# r.text

import requests

response=requests.get('https://api.github.com/events')

response.text
```

```python
# json.loads(r.text)

import json

json.loads(response.text)
```
