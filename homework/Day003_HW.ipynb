### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
# File I/O

fh=open('./data/64_72hr_CH.xml','r',encoding='utf-8')
xml=fh.read()
fh.close()
print(xml)

# xmltodict


import xmltodict

with open("./data/64_72hr_CH.xml",encoding='utf-8') as fd:
    xml=fd.read()
    doc=dict(xmltodict.parse(xml))

print(doc)

# File I/O 印出的結果會含有許多標籤
# xmltodict則不會有許多標籤
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict

with open("./data/64_72hr_CH.xml",encoding='utf-8') as fd:
    doc=dict(xmltodict.parse(fd.read()))

data=doc['cwbopendata']['dataset']['locations']['location']

print(len(data))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

import xmltodict

with open("./data/64_72hr_CH.xml",encoding='utf-8') as fd:
    doc=dict(xmltodict.parse(fd.read()))

data=doc['cwbopendata']['dataset']['locations']['location']

for i in range(len(data)):
    print('地區:',data[i]['locationName'],'第一個時間點:',data[i]['weatherElement'][0]['time'][0]['dataTime'],'溫度:',data[i]['weatherElement'][0]['time'][0]['elementValue']['value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

import xmltodict

with open("./data/64_72hr_CH.xml",encoding='utf-8') as fd:
    doc=dict(xmltodict.parse(fd.read()))

data=doc['cwbopendata']['dataset']['locations']['location']

print('地區:',data[0]['locationName'])
for i in range(len(data[0]['weatherElement'][0]['time'])):
    print('時間點:',data[0]['weatherElement'][0]['time'][i]['dataTime'],'溫度:',data[0]['weatherElement'][0]['time'][i]['elementValue']['value'])
```
