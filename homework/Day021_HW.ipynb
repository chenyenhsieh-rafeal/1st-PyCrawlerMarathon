## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from datetime import timedelta, datetime
import time
```

```python
url="https://www.ettoday.net/news/news-list.htm"

# 得到今天日期
formatted_today = time.strftime("%Y/%m/%d")

browser=webdriver.Chrome(executable_path="chromedriver")

# 進入網站
browser.get("https://www.ettoday.net/news/news-list.htm")

# 自動下滑直到今日新聞爬取完畢
keepGoing=True
while keepGoing:
    time.sleep(2)
    browser.execute_script("window.scrollBy(0,10000)")
    html_source=browser.page_source
    soup=BeautifulSoup(html_source,"lxml")
    
    item_list=soup.find(name="div",attrs={"class":"part_list_2"}).find_all('h3')
    
    for item in item_list:
        if str(item.span.text[:10])==formatted_today:
            keepGoing=True
        else:
            keepGoing=False

# 解析網頁
html_source=browser.page_source
soup=BeautifulSoup(html_source,"lxml")

item_list=soup.find(name="div",attrs={"class":"part_list_2"}).find_all('h3')

for item in item_list:
    if str(item.span.text[:10])==formatted_today:
        print(item.span.text,item.em.text,item.a.text)
    else:
        break

browser.quit()
```

## 2. 取出現在時間兩小時內的新聞

```python
url="https://www.ettoday.net/news/news-list.htm"

# 得到今天日期
formatted_today = time.strftime("%Y/%m/%d")

browser=webdriver.Chrome(executable_path="chromedriver")

# 進入網站
browser.get("https://www.ettoday.net/news/news-list.htm")

# 自動下滑直到今日新聞爬取完畢
keepGoing=True
while keepGoing:
    time.sleep(2)
    browser.execute_script("window.scrollBy(0,10000)")
    html_source=browser.page_source
    soup=BeautifulSoup(html_source,"lxml")
    
    item_list=soup.find(name="div",attrs={"class":"part_list_2"}).find_all('h3')
    
    for item in item_list:
        if str(item.span.text[:10])==formatted_today:
            keepGoing=True
        else:
            keepGoing=False

# 解析網頁
for item in item_list:
    if str(item.span.text[:10])==formatted_today and int(item.span.text[11:13])>=21:
        print(item.span.text,item.em.text,item.a.text)
        
browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select
from datetime import timedelta, datetime

three_age = datetime.today() + timedelta(-3)
formatted_day = three_age.strftime('%Y/%m/%d')

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

selectYear=Select(browser.find_element_by_id("selY"))
selectYear.select_by_value(formatted_day[0:4])
selectMonth=Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value(formatted_day[5:7])
selectDay=Select(browser.find_element_by_id("selD"))
selectDay.select_by_value(formatted_day[8:10])

browser.find_element_by_id('button').click()

# 設定判斷控制器，僅取得當日擷取範圍
keepGoing = True
while keepGoing:
    time.sleep(2)
    browser.execute_script("window.scrollBy(0,10000)")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "lxml")

    div_find_h3 = soup.find(name="div",attrs={"class":"part_list_2"}).find_all("h3")
    
    for i in div_find_h3:
        if (str(i.span.text[:10]) == formatted_day):
            keepGoing = True
        else:
            keepGoing = False

# 解析相關網頁內容
html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")

for i in div_find_h3:
    if (int(i.span.text[11:13])>=15 and int(i.span.text[11:13])<=17):
        print(i.span.text,i.em.text,i.a.text)
        
browser.quit()
```
