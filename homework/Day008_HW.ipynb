### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests

url = 'https://www.dcard.tw/f'

r=requests.get(url)
r.encoding = 'utf-8'

response=r.text

print(response[0:3000])
```

```python
# 1. (1) Request 取回之後該怎麼取出資料
#    (2) 資料型態是什麼？

# (1) Request取回後用 r=requests.get(url),response=r.text,response即為欲取出的資料 
# (2) 資料型態是 str
print(type(response))
```

```python
# 2. (1) 為什麼要使用 BeatifulSoup 處理？
#    (2) 處理後的型態是什麼？

# (1) 用BeautifulSoup處理可以讓我們解析由 requests取得的資料,也可以利用BeautifulSoup取得或是挑出我們想要尋找的特定資料
# (2) 經BeautifulSoup處理過的資料型態是 bs4.BeautifulSoup
from bs4 import BeautifulSoup

soup=BeautifulSoup(response)

print(type(soup))
```

### 2. 知乎： https://www.zhihu.com/explore

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'

print(r.text[0:600])
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
# 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

# 一開始因為是用最直接的方式進入知乎爬取資料,被知乎認定為非尋常的瀏覽,所以無法爬取資料,
# 後來用了headers偽裝成一般瀏覽器的正常瀏覽後,即可成功爬取資料

import requests

url = 'https://www.zhihu.com/explore'

headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'}

r = requests.get(url,headers=headers)
r.encoding = 'utf-8'

print(r.text)
```
